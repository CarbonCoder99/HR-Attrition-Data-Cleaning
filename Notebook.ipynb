```python
data = pd.read_csv('RawData/HR-Employee-Attrition.csv')
```

```python
## TASK 1 - Delete Redundant columns
columns_to_keep = ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department', 'DistanceFromHome', 'Education', 'HourlyRate', 'JobRole', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
columns_to_drop = ['EducationField', 'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'Over18', 'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance']
```

```python
len(columns_to_keep)
```

```python
len(columns_to_drop)
```

```python
# First method of dropping columns - Filter columns to keep
data_reduced_1 = data[columns_to_keep]
```

```python
data_reduced_1
```

```python
# Second method of dropping columns - Drop columns to drop
data = data.drop(columns=columns_to_drop)
```

```python
# This confirms that the redundant columns have been removed
data
```

```python
# TASK 2 - Rename columns to remove spaces and make them more intuitive
data = data.rename(columns={
    'Age': 'age',
    'Attrition': 'attrition',
    'BusinessTravel': 'business_travel',
    'DailyRate': 'daily_rate',
    'Department': 'department',
    'DistanceFromHome': 'distance_from_home',
    'Education': 'education',
    'HourlyRate': 'hourly_rate',
    'JobRole': 'job_role',
    'YearsAtCompany': 'years_at_company',
    'YearsInCurrentRole': 'years_in_current_role',
    'YearsSinceLastPromotion': 'years_since_last_promotion',
    'YearsWithCurrManager': 'years_with_curr_manager'
})
```

```python
# This confirms that the columns have been renamed
data.columns
```

```python
# TASK 3 - Drop duplicate rows if any
data.duplicated().sum()
# This confirms that there are no duplicate rows in the dataset
```

```python
# TASK 4 - Check for missing values
data.isna().sum()
# This confirms that there are no missing values in the dataset
```

```python
# TASK 5 - For the sake of analysis, we only want to keep employees who have been with the company for nothing more than 15 years
data = data[data['years_at_company'] <= 8]
```

```python
data.shape
```

```python
# TASK 6 - Let's clean the business_travel column to make it more intuitive
data['business_travel'] = data['business_travel'].str.replace("_", " ").str.lower()
```

```python
data
```

```python
# TASK 7 - Say we want to feed our data into a machine learning model, we need to convert categorical columns to numerical columns. Let's do that for the attrition column
data['attrition'] = data['attrition'].map({'Yes': 1, 'No': 0})
# we then convert the attrition column to boolean type
data['attrition'] = data['attrition'].astype(bool)
```

```python
data
```

```python
# FINAL TASK - Save the cleaned data to a new CSV file
data.to_csv('ProcessedData/HR-Employee-Attrition-Cleaned.csv', index=False)
```
